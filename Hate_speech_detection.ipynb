```python
dataset["label"] = dataset["class"].map({0:"Hate speech",
                                         1:"Offensive Language",
                                         2:"No hate or offensive"})
dataset
```

```python
data = dataset[["tweet","label"]]
```

```python
data
```

```python
#importing regular expression and natural language toolkit
import re
import nltk
import string
nltk.download('stopwords')
#removal of stop words and stemming the words
```

```python
#importing of stop words
from nltk.corpus import stopwords
stopwords = set(stopwords.words("english"))
```

```python
#importing stemming
stemmer = nltk.SnowballStemmer("english")
```

```python
#DataCleaning
def clean_data(text):
    text = str(text).lower()
    text = re.sub('https?://\s+www\.S+'," ",text)
    text = re.sub('\[.*?\]',' ',text)
    text = re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text = re.sub('\n','',text)
    text = re.sub('\w*\d\w*', '',text)
    #stopwords removal
    text = [word for word in text.split(' ') if word not in stopwords]
    text = " ".join(text)
    #stemming the text
    text =[ stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text
    
```

```python
data["tweet"] = data["tweet"].apply(clean_data)
```

```python
data
```

```python
X = np.array(data["tweet"])
Y = np.array(data["label"])
```

```python
X
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv = CountVectorizer()
X = cv.fit_transform(X)
```

```python
X
```

```python
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
#Building ML model
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
#confusion matrix and accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.heatmap(cm, annot = True, fmt="d" , cmap="YlGnBu" )
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```
